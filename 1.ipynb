Data Visualization is simply graphical representation of data or any information.
That's where matplotlib is used.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   # just importing a pyplot from matplotlib 
print("libraries imported")
```

Line Plot Chart :- Line plots are excellent for displaying trends or patterns over a continuous range. For example, you might use a line plot to show how a stock price changes over time.

```python
country_population = [700,750,740,730,752,800,790]
country_population
```

```python
# lineplot()

plt.plot(country_population)
```

```python
plt.plot(country_population)
plt.show()
```

```python
# let's consider 2d line chart
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population)
plt.show()
```

```python
# marking indicator point
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population,marker="*",ms=15,mfc='yellow')  # ms = marker size, mfc = marker face color
plt.show()
```

```python
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population,marker="D",ms=15,mfc='yellow')  # ms = marker size, mfc = marker face color
# marker = "*", "v", "D", "p", "P" ....
plt.show()
```

```python
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population,marker="v",ms=15,mfc='yellow')  # ms = marker size, mfc = marker face color
plt.show()
```

label representation on axis

```python
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population,marker="p",ms=15,mfc='yellow')  # ms = marker size, mfc = marker face color
plt.xlabel("Axis-X")
plt.ylabel("country_population")
plt.show()
```

```python
list1 = [100,400,300,400,100]
list2 = [1,1,2,3,3]
plt.plot(list2,list1)
plt.show()
```

Changing color of line and give title to graph and ls parameter

```python
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population,marker="*",ms=15,mfc='yellow',color='blue',ls = 'dashed',lw=4)  # ms = marker size, mfc = marker face color
plt.title("Lineplot Demo")
plt.xlabel("Axis-X")
plt.ylabel("country_population")
plt.show()
# ls = line style = dotted , dashed, solid(by default)
# lw = line width
```

BAR PLOT :- 

```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']

plt.bar(country_name,country_population)
plt.xlabel('name')
plt.ylabel('Population')
plt.show()
```

```python
df = pd.read_csv('Used_Bikes.csv')
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df['brand'].value_counts()
```

```python
name_of_brand = list(df['brand'].value_counts().keys())
name_of_brand
```

```python
no_of_values = list(df['brand'].value_counts().values)
no_of_values
```

```python
plt.bar(name_of_brand,no_of_values)
plt.show()
```

```python
# Removing overlapping of brand names in above diagram using rotation of ticks of axis x and y
plt.bar(name_of_brand,no_of_values)
plt.xticks(rotation = 80)
plt.show()
```

Designing bar graph

```python
plt.figure(figsize=(10,4))  #(width,height)
plt.bar(name_of_brand,no_of_values,color='red',ec='black')
plt.xticks(rotation = 80)

# ec = edge color

plt.savefig('brandpopulation')  # save graph image with name of brandpopulation

plt.show()
```
